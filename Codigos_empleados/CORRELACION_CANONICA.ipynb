```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import fft, fftfreq
from scipy import signal
from scipy.signal import butter, lfilter
```

```python
def butter_bandpass(lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='bandpass')
    return b, a


def butter_bandpass_filter(data, lowcut, highcut, fs, order=5):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = lfilter(b, a, data)
    return y

def org_senal(senal):
    temp = []
    for fila in senal:
        for dato in fila:
            temp.append(dato)

    vector_fila = np.array(temp).reshape(1, -1)
    vector_fila

    z = []
    for fil in vector_fila:
        for dat in fil:
            z.append(dat)
            
    z_np = np.transpose(z)
    
    return z_np

def crea_senal(fs,n_samples,freq):

    duration = n_samples / fs
    time = np.linspace(0, duration, n_samples, endpoint=False)

    signal = np.sin(2 * np.pi * freq * time)
    
    return signal

def correlacion(signal1,signal2):
    # Calcular la correlación cruzada normalizada de las señales
    corr = np.correlate(signal1, signal2) / np.correlate(signal1, signal1,)

    # El índice de correlación es el valor máximo de la correlación cruzada normalizada
    corr_index = np.max(corr)
    
    return corr_index

def Valores_correlacion(sen):

    senal_10_hz = pd.read_csv("../CODIGOS_FINALES/SUJETOS/S1/arriba",sep=';')
    senal_6_6_hz = pd.read_csv("../CODIGOS_FINALES/SUJETOS/S1/derecha",sep=';')
    senal_7_5_hz = pd.read_csv("../CODIGOS_FINALES/SUJETOS/S1/abajo",sep=';')
    senal_8_5_hz = pd.read_csv("../CODIGOS_FINALES/SUJETOS/S1/izquierda",sep=';')

    fs = 250
    lowcut = 0.4
    highcut = 30
    sen10 = butter_bandpass_filter(senal_10_hz , lowcut, highcut, fs,order=6)
    sen6_6 = butter_bandpass_filter(senal_6_6_hz , lowcut, highcut, fs,order=6)
    sen7_5 = butter_bandpass_filter(senal_7_5_hz , lowcut, highcut, fs,order=6)
    sen8_5 = butter_bandpass_filter(senal_8_5_hz , lowcut, highcut, fs,order=6)



    #f1
    lowcut1 = 5.6
    highcut1 = 7.6
    senf1 = butter_bandpass_filter(sen6_6 , lowcut1, highcut1, fs,order=6)
    senf1 = org_senal(senf1)

    #f2
    lowcut2 = 6.5
    highcut2 = 8.5
    senf2 = butter_bandpass_filter(sen7_5 , lowcut2, highcut2, fs,order=6)
    senf2 = org_senal(senf2)

    #f3
    lowcut3 = 7.9
    highcut3 = 9.5
    senf3 = butter_bandpass_filter(sen8_5 , lowcut3, highcut3, fs,order=6)
    senf3 = org_senal(senf3)

    #f4
    lowcut4 = 9
    highcut4 = 11
    senf4 = butter_bandpass_filter(sen10 , lowcut4, highcut4, fs,order=6)
    senf4 = org_senal(senf4)

    
    fs = 250
    lowcut = 0.4
    highcut = 30
    senf10 = butter_bandpass_filter(sen , lowcut, highcut, fs,order=6)


    #f1
    lowcut1 = 5.6
    highcut1 = 7.6
    senf11 = butter_bandpass_filter(senf10 , lowcut1, highcut1, fs,order=6)
    senf11 = org_senal(senf11)

    #f2
    lowcut2 = 6.5
    highcut2 = 8.5
    senf21 = butter_bandpass_filter(senf10 , lowcut2, highcut2, fs,order=6)
    senf21 = org_senal(senf21)

    #f3
    lowcut3 = 7.9
    highcut3 = 9.5
    senf31 = butter_bandpass_filter(senf10 , lowcut3, highcut3, fs,order=6)
    senf31 = org_senal(senf31)

    #f4
    lowcut4 = 9
    highcut4 = 11
    senf41 = butter_bandpass_filter(senf10 , lowcut4, highcut4, fs,order=6)
    senf41 = org_senal(senf41)

    Corr1 = correlacion(signal1=senf11,signal2=senf4)
    Corr2 = correlacion(signal1=senf21,signal2=senf4)
    Corr3 = correlacion(signal1=senf31,signal2=senf4)
    Corr4 = correlacion(signal1=senf41,signal2=senf4)

    COR1 = [Corr1,Corr2,Corr3,Corr4]
    max1 = np.max(COR1)

    Corr11 = correlacion(signal1=senf11,signal2=senf1)
    Corr21 = correlacion(signal1=senf21,signal2=senf1)
    Corr31 = correlacion(signal1=senf31,signal2=senf1)
    Corr41 = correlacion(signal1=senf41,signal2=senf1)

    COR2 = [Corr11,Corr21,Corr31,Corr41]
    max2 = np.max(COR2)

    Corr12 = correlacion(signal1=senf11,signal2=senf2)
    Corr22 = correlacion(signal1=senf21,signal2=senf2)
    Corr32 = correlacion(signal1=senf31,signal2=senf2)
    Corr42 = correlacion(signal1=senf41,signal2=senf2)

    COR3 = [Corr12,Corr22,Corr32,Corr42]
    max3 = np.max(COR3)

    Corr13 = correlacion(signal1=senf11,signal2=senf3)
    Corr23 = correlacion(signal1=senf21,signal2=senf3)
    Corr33 = correlacion(signal1=senf31,signal2=senf3)
    Corr43 = correlacion(signal1=senf41,signal2=senf3)

    COR4 = [Corr13,Corr23,Corr33,Corr43]
    max4 = np.max(COR4)

    totalmax = [max1,max2,max3,max4]
    
    return totalmax
```
